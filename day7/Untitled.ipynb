```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,StratifiedKFold,GridSearchCV
from sklearn.metrics import log_loss, confusion_matrix,classification_report,ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder
```

```python
img = pd.read_csv(r"F:\PML\Cases\Image_Segmentation\Image_Segmention.csv")
le = LabelEncoder()
img['Class'] = le.fit_transform(img['Class'])
```

```python
X = img.drop('Class', axis=1)
y = img['Class']
```

```python
params = {'penalty':[None, 'l1', 'l2', 'elasticnet'],
        'C' : np.linspace(0,10,10),
        'l1_ratio': np.linspace(0,10,5),
         'multi_class' : ['ovr', 'multinomial']}
lr = LogisticRegression(random_state = 23, solver='saga')
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state = 23)
gcv = GridSearchCV(lr, param_grid=params, cv=kfold, scoring='neg_log_loss')
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```

#### inferencing

```python
#inferencing on the unlabelled data
bm = gcv.best_estimator_
tst = pd.read_csv(r"F:\PML\Cases\Image_Segmentation\tst_img.csv")
print(bm.predict(tst))
print(dict(zip(le.classes_, np.arange(0,7))))
```

```python
prediction = le.inverse_transform(bm.predict(tst))
print(prediction)
```
