```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression, ElasticNet
from sklearn.model_selection import train_test_split,StratifiedKFold, GridSearchCV
from sklearn.metrics import log_loss
```

```python
kyphosis = pd.read_csv(r"F:\PML\Cases\Kyphosis\Kyphosis.csv")
kyphosis.head()
#kyphosis.isnull().sum()
```

```python
X = kyphosis.drop('Kyphosis', axis=1)
y = kyphosis['Kyphosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=23,stratify=y)
```

#### logloss

```python
lr = LogisticRegression(penalty='l1', solver='saga', l1_ratio=0.5)
lr.fit(X_train, y_train)
y_pred_proba = lr.predict_log_proba(X_test)
#print(y_pred_proba)
print(log_loss(y_test, y_pred_proba))
```

```python
params = {'penalty':[None, 'l1', 'l2', 'elasticnet'],
        'C' : np.linspace(0,10,5),
        'l1_ratio': np.linspace(0,1,10)}
```

```python
lr = LogisticRegression(solver='saga')
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state = 23)
```

```python
gcv = GridSearchCV(lr, param_grid=params, cv=kfold, scoring='neg_log_loss')
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
pd_cv = pd.DataFrame(gcv.cv_results_)
print(pd_cv)
```
